```python
trial_train = train_file_1[["X_LOC","Y_LOC","Z_LOC","CALI","DEPTH_MD","RDEP", "RHOB", "SGR", 
                "NPHI", "PEF", "DTC","SP","BS","ROP","FORCE_2020_LITHOFACIES_LITHOLOGY"]].copy()
```

```python
trial_train.head(9)
```

```python
trial_train.rename(columns={'FORCE_2020_LITHOFACIES_LITHOLOGY':'NPD_LITHO'}, inplace=True)
```

```python
trial_train.head(9)
```

```python
lithology_numbers = {30000: 'Sandstone',
                 65030: 'Sandstone/Shale',
                 65000: 'Shale',
                 80000: 'Marl',
                 74000: 'Dolomite',
                 70000: 'Limestone',
                 70032: 'Chalk',
                 88000: 'Halite',
                 86000: 'Anhydrite',
                 99000: 'Tuff',
                 90000: 'Coal',
                 93000: 'Basement'}
```

```python
simple_lithology_numbers = {30000: 1,
                 65030: 2,
                 65000: 3,
                 80000: 4,
                 74000: 5,
                 70000: 6,
                 70032: 7,
                 88000: 8,
                 86000: 9,
                 99000: 10,
                 90000: 11,
                 93000: 12}
```

```python
trial_train['LITH'] = trial_train['NPD_LITHO'].map(lithology_numbers)
trial_train['LITH_SI'] = trial_train['NPD_LITHO'].map(simple_lithology_numbers)
```

```python
trial_train.head(9)
```

```python
g = sns.FacetGrid(trial_train, col='LITH', col_wrap=4)
g.map(sns.scatterplot, 'NPHI', 'RHOB', alpha=0.5)
g.set(xlim=(-0.15, 1))
g.set(ylim=(3, 1))
```

```python
trial_train_x = trial_train.drop(["NPD_LITHO","LITH","LITH_SI"],axis = 1)
trial_train_y = trial_train["LITH_SI"]
```

```python
trial_train_x.head()
```

```python
trial_train_y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(trial_train_x,trial_train_y, test_size=0.2, train_size=0.8)
x_train_1,x_test_1,y_train_1,y_test_1 = train_test_split(x_train,y_train, test_size=0.2, train_size=0.8)
```

```python
model_1 = xg.XGBClassifier(tree_method = 'gpu_hist')
model_1.fit(x_train_1,y_train_1)
```

```python
y_pred_1 = model_1.predict(x_test_1)
sklearn.metrics.accuracy_score(y_test_1,y_pred_1)
```
